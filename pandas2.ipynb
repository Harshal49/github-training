```python
st_details={'st_id':[1,2,3,4,5],
           'st_name':['a','b','c','d','e'],
           'region':['n','d','m','e','mp']}


st_marks={'st_id':[1,2,3,4,6],
           'sub':['py','ds','ml','tab','sql'],
           'marks':[35,32,34,37,38]}
```

```python
st_details
```

```python
import pandas as pd
```

```python
pd.__version__
```

```python
df=pd.DataFrame(st_details)
```

```python
df
```

```python
type(df)
```

```python
df1=pd.DataFrame(st_marks)
```

```python
df1
```

```python
pd.merge(df,df1,on='st_id',how='left')
```

```python
pd.merge(df,df1,on='st_id',how='outer')
```

```python
#when the column name are different then

pd.merge(df,df1,left_on='st_id',right_on='st_id',how='outer')
```

```python
df3=pd.concat([df,df1],axis=0)   #(axis1 for column,axis=0 for  rows)
```
