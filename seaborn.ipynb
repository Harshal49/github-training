# jointplot

```python
import seaborn as sns
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

# pairplot

pairplot isalsoknown as scatter plot.in which one variable in the same data row is matched with another variable's value.pair plot is just elabrations on this ,showing all variables paired with all the other variables.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

# Dist plot

```python
# distplot is use for histogram
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

# categorical plot
for categorical data
1.boxplot
2.violinplot
3.countplot
4.bar plot

```python
#count plot

sns.countplot('sex',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
## bar plot

sns.barplot(y='total_bill',x='sex',data=df)
```

# box plot

```python
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

# violin plot

```python
#display the distribution

sns.violinplot(x='total_bill',y='day',data=df)
```
